The Waist-Hip Ratio is defined by using the variables W for weight and H for height. This ratio shows the relationship between the circumfrence of the waist and the hips. This is calculated by dividing W by H (W/H=WHR).

## Source Data 

<!--- Paste in your data definition table from part 2. --->
- ID : Identifiaction number for each user.
- Waist : Waist value (inches) for each user.
- Hips : Hip value (inches) for each user.
- Gender : Gender for each user (M or F).

## Data Import

```python
# Goal: Extract the data from the file
# import pandas as pd & numpy as np
import numpy as np
import pandas as pd

# import W2h_data.csv file and check data types are assigned correctly
rows = pd.read_csv('w2h_data.csv',dtype={"ID":np.int64, "Waist":np.float64,"Hip": np.float64, "Gender":np.str})
#rows.head()
rows.dtypes
```

## Calculations

```python
# Goal: For each row of data calculate and store the w2h_ratio and shape.

# calculate the waist to hips ratios
w2h_ratios = rows['Waist']/rows['Hip']

# determine who is male (True) or female (False)
males = (rows['Gender']=='M')

# males with w2h ration > 0.9 are apple shaped
male_apples = np.logical_and((w2h_ratios>0.9),males)

# females with w2h ratio > 0.8 are apple shaped
female_apples = np.logical_and((w2h_ratios>0.8),np.logical_not(males))

# combines the male_apples and female_apples
apples=np.logical_or(male_apples,female_apples)

# works like an if then else statement
shapes=np.where(apples,'Apple','Pear')

# The code below creates a new results array 
    
# add the two new columns
rows['W2H Ratio']=w2h_ratios
rows['Shape']=shapes
rows

# convert to dataframe and delete ID column
bmi = pd.DataFrame(rows, index=rows.ID -1)
del bmi['ID']
bmi
```

## Output

```python
# Goal: pretty print the rows as an HTML table

# Convert to HTML table
bmi.to_html("bmi.html")

# Export to a CSV called "StatsResults.csv"
bmi.to_csv("StatsResults.csv")
```
